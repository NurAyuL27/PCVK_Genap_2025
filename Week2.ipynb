# **4. Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image dengan tipe data int16.**

```python
black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

**Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2**

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_img, pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
```

**Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2 yang tertulis pada kode program.**

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

**Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang
tertulis pada kode program.**

```python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

**Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut**

```python
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

**Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah
ditentukan.**

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello',org=(18,500),fontFace=font, fontScale=4, color=(255,255,0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

**Pembuatan black image kembali dilakukan dengan tipe data int32**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random

# --- Baca gambar dari Google Drive ---
img = cv2.imread("/content/drive/MyDrive/kitten.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

**1. Dengan menggunakan figsize, perhatikan apakah ukuran image pixelnya juga
berubah?**

```python
plt.figure(figsize=(4,4))
plt.imshow(img_rgb)
plt.title("Figsize 4x4")
plt.axis("off")

plt.figure(figsize=(10,10))
plt.imshow(img_rgb)
plt.title("Figsize 10x10")
plt.axis("off")
plt.show()
print("Ukuran pixel asli:", img_rgb.shape)
```

**2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!**

```python
rb = img_rgb.copy()
rb[:,:,1] = 0   # hilangkan Green

gb = img_rgb.copy()
gb[:,:,0] = 0   # hilangkan Red

plt.subplot(1,2,1)
plt.imshow(rb)
plt.title("Red-Blue")

plt.subplot(1,2,2)
plt.imshow(gb)
plt.title("Green-Blue")
plt.show()
```

**3. Tampilkan image baris ke 20-115, kolom 25-120!**

```python
crop1 = img_rgb[20:116, 25:121, :]
plt.imshow(crop1)
plt.title("Cropped Rows 20–115, Cols 25–120")
plt.axis("off")
plt.show()
```

**4. Tampilkan image baris ke 5-30, semua kolom, channel Red saja!**

```python
red_only = img_rgb[5:31, :, 0]
plt.imshow(red_only, cmap="Reds")
plt.title("Rows 5–30, Red channel")
plt.axis("off")
plt.show()
```

**5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
disarankan menggunakan bilangan acak/random!**

```python
canvas = np.zeros((400,400,3), dtype=np.uint8)
for _ in range(5):
    pt1 = (random.randint(0,300), random.randint(0,300))
    pt2 = (pt1[0]+random.randint(20,100), pt1[1]+random.randint(20,100))
    color = (random.randint(0,255), random.randint(0,255), random.randint(0,255))
    cv2.rectangle(canvas, pt1, pt2, color, -1)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.title("5 Kotak Random")
plt.axis("off")
plt.show()
```

**6. Tampilkan image dengan posisi terbalik!**

```python
flipped = cv2.flip(img_rgb, 0)
plt.imshow(flipped)
plt.title("Flipped Image (Vertical)")
plt.axis("off")
plt.show()
```

**7. Buat rectangle dan circle pada bagian wajah dari image foto anda saat
beraktifitas (bukan pasfoto).**

```python
# Gunakan Haarcascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalcatface.xml')

gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.1, 5)

img2 = img_rgb.copy()
for (x,y,w,h) in faces:
    cv2.rectangle(img2, (x,y), (x+w,y+h), (255,0,0), 3)
    cv2.circle(img2, (x+w//2, y+h//2), min(w,h)//2, (0,255,0), 3)

plt.imshow(img2)
plt.title("Rectangle + Circle di Wajah Kucing (Auto Detect)")
plt.axis("off")
plt.show()
```

**8. Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB
dari citra kitten/ lena/ mandrill/ male/ female/ couple/ sailboat/ peppers!**

```python
blue_channel = img[:,:,0]
h, w = blue_channel.shape
roi = blue_channel.copy()
cv2.rectangle(roi, (0,h-50), (50,h), 0, -1)   # rectangle hitam

plt.imshow(roi, cmap="gray")
plt.title("Rectangle di sudut kiri bawah (channel B)")
plt.axis("off")
plt.show()
```
